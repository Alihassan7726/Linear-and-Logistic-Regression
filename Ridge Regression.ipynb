```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# dataset
from sklearn.datasets import load_boston
# scaling and dataset split
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
# OLS, Ridge
from sklearn.linear_model import LinearRegression, Ridge
# model evaluation
from sklearn.metrics import r2_score, mean_squared_error
```

```python
house = load_boston()
df = pd.DataFrame(house.data , columns = house.feature_names)

df['Price'] = house.target
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import minmax_scale
df[house.feature_names.tolist()] = minmax_scale(df[house.feature_names.tolist()])
df[house.feature_names.tolist()]
```

```python
X = df[house.feature_names.tolist()]
y = df['Price']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.1 , random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import cross_val_score
ln = LinearRegression()
accuracies = cross_val_score(estimator = ln, X = X_train,\
     y = y_train, cv = 10)
print("Accuracy Mean {} Accuracy Variance \
     {}".format(accuracies.mean(),accuracies.std()))
print("\n",accuracies)
```

```python
from sklearn.model_selection import cross_val_score 
mse = cross_val_score(estimator = ln, X = X_train,\
     y = y_train, cv = 10 , scoring = 'neg_mean_squared_error')
print("Error Mean {} Error Variance \
     {}".format(mse.mean(),mse.std()))
print("\n",mse)
```

```python
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha': [0.001,0.0001,0.005,np.arange(1,100,0.5)]}

ridge_reg = GridSearchCV(estimator = ridge , param_grid = params , scoring = 'neg_mean_squared_error' , cv = 10)
ridge_reg.fit(X_train , y_train)
```

```python
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```
